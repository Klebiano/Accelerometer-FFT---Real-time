```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}
plt.rcParams.update({'font.size': 13, 'font.weight':'bold', 'figure.figsize': (10,6)})
```

```python
Fs = 2000
t = 1/Fs
x = np.linspace(0, 2*np.pi, 500)
func = 2.0*np.sin(x*2*np.pi) + 3.0*np.sin(x*2*np.pi*100.0)
```

```python
n = len(func)
k = np.arange(n)
T = n/Fs
frq = k/T 
frq = frq[range(n//2)]
```

```python
Y = np.fft.fft(func)/n
Y = Y[range(n//2)]
```

```python
fig, ax = plt.subplots(2, 1)
plt.subplots_adjust(hspace=0.4)
ax[0].plot(x,func)
ax[0].set_xlabel('Time', fontweight='bold')
ax[0].set_ylabel('Amplitude', fontweight='bold', labelpad=20)
ax[1].plot(frq,abs(Y),'r')         
ax[1].set_xlabel('Freq (Hz)', fontweight='bold')
ax[1].set_ylabel('|Y(freq)|', fontweight='bold', labelpad=20)
ax[0].set_title('Dados acelermetro', fontweight='bold');
```
